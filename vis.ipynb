```python
data = [
    [0.287, 0.391],
    [0.409, 0.402]
]
std = [
    [0.247, 0.282],
    [0.271, 0.278]
]
```

```python
fig, ax = plt.subplots(figsize=(3,4))

X = np.arange(2)
width = 0.25

ax.bar(X - width/2, data[0], yerr=std[0], 
    width = width, capsize=5, label='Darker')
ax.bar(X + width/2 + 0.05, data[1], yerr=std[1], 
    width = width, capsize=5, label='Lighter')

ax.set_xticks(X)
ax.set_xticklabels(['No Att.', 'Att.'])
ax.set_title('Pass Rate ($\\uparrow$)')
ax.legend(loc='upper left', framealpha=True)
fig.savefig('pass_rate.png', dpi=300)
```
